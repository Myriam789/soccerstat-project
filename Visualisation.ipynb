```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('top5-players_CLEANED.csv')

```

```python
positions = data['Pos'].dropna().str.split(',')
all_positions = [p.strip() for group in positions for p in group]
position_counts = pd.Series(all_positions).value_counts()

```

```python
plt.figure(figsize=(8,5))
position_counts.plot(kind='bar', color='skyblue')
plt.title('Players by Position')
plt.xlabel('Position')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
nation_counts = data['Nation'].value_counts().head(10)
```

```python
plt.figure(figsize=(8,5))
nation_counts.plot(kind='bar', color='lightgreen')
plt.title('Players by Nation (Top 10)')
plt.xlabel('Nation')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
