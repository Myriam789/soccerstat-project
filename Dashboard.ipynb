```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('top5-players_CLEANED.csv')
```

```python
position_counts = df['Pos'].value_counts()

position_counts_sorted = position_counts.sort_values(ascending=False)

print("Player Count by Position")
print(position_counts_sorted)
```

```python
plt.figure(figsize=(10, 6))
position_counts_sorted.plot(kind='bar', color='#8a29d9ff')

plt.title('Number of Players by Position ', fontsize=16)
plt.xlabel('Position', fontsize=12)
plt.ylabel('Number of Players', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()


plt.savefig('players_by_position_bar_chart.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('top5-players_CLEANED.csv')
```

```python
nation_counts = df['Nation'].value_counts()

top_nations = nation_counts.head(15)

print(" Player Count by Top 15 Nation")
print(top_nations)
```

```python
plt.figure(figsize=(12, 7))
top_nations.plot(kind='bar', color="#ffef41")
plt.title('Top 15 Nations by Number of Players (Min >= 450)', fontsize=16)
plt.xlabel('Nation', fontsize=12)
plt.ylabel('Number of Players', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('players_by_nation_bar_chart.png')
```
